1. 데이터 전처리 (데이터 전처리)
  1.1. 결측치 처리
  결측값 확인 및 대체: 각 변수에서 결측치가 있는지 확인하고, 적절한 방법(평균, 중앙값, 최빈값 등)으로 대체하거나 해당 행/열을 제거합니다.
  특수 값 처리: 'Unknown', 'N/A' 등과 같은 특수 문자를 결측치로 인식하고 처리합니다.
  1.2. 이상치 탐지 및 처리
  이상치 탐지: 박스 플롯, IQR(Interquartile Range) 등을 사용하여 이상치를 식별합니다.
  이상치 처리: 이상치를 제거하거나, 다른 값으로 대체(예: 클리핑)합니다.
  1.3. 데이터 정규화 및 표준화
  정규화: 연속형 변수의 범위를 0과 1 사이로 조정하여 비교가 용이하게 만듭니다.
  표준화: 평균을 0, 표준편차를 1로 조정하여 변수 간의 스케일 차이를 줄입니다.
  1.4. 범주형 변수 인코딩
  라벨 인코딩: 범주형 데이터를 숫자로 변환합니다.
  원-핫 인코딩: 범주형 변수를 더미 변수로 변환하여 모델링에 활용합니다.
  1.5. 파생 변수 생성
  선수 연령 계산: 출생 연도와 대회 연도를 이용해 선수의 연령을 계산합니다.
  경력 기간: 선수의 첫 대회 참가 연도와 마지막 대회 연도를 이용해 경력 기간을 계산합니다.
  1.6. 날짜/시간 데이터 처리
  대회 날짜: 대회의 개최 연도, 시즌(여름/동계) 등을 추출하여 분석에 활용합니다.
  경기 일정: 특정 경기의 날짜와 시간을 분석에 반영할 수 있습니다.
  1.7. 데이터 통합
  다양한 데이터 소스 병합: 선수 정보, 국가 정보, 경기 정보 등을 통합하여 종합적인 분석이 가능하도록 합니다.

2. 데이터 시각화 (데이터 시각화)
  2.1. 국가별 메달 획득 추이
  시계열 그래프: 연도별로 국가별 메달 획득 추이를 시각화하여 강국의 변화를 파악합니다.
  스택형 막대 그래프: 각 국가의 금, 은, 동 메달 수를 누적하여 비교합니다.
  2.2. 메달 종류별 분포
  도넛 차트 또는 파이 차트: 금, 은, 동 메달의 비율을 시각화합니다.
  막대 그래프: 각 메달 종류별로 획득한 수량을 비교합니다.
  2.3. 스포츠 종목별 인기 및 경쟁도
  막대 그래프: 각 스포츠 종목별로 참가 선수 수나 메달 수를 시각화합니다.
  히트맵: 종목과 국가 간의 메달 획득 관계를 색상으로 표현합니다.
  2.4. 선수별 기록 비교
  산점도: 선수의 기록(예: 속도, 점수 등)을 두 변수 간의 관계로 시각화합니다.
  레이더 차트: 여러 변수에서의 선수 성과를 종합적으로 비교합니다.
  2.5. 인구통계적 정보 시각화
  히스토그램: 선수의 연령 분포를 시각화합니다.
  파이 차트: 성별 비율을 보여줍니다.
  바이올린 플롯: 국가별 선수 연령 분포를 시각화합니다.
  2.6. 지리적 분포
  지도 시각화(Geo Map): 국가별 메달 획득 수를 지도 위에 표시하여 지리적 패턴을 파악합니다.
  버블 맵: 국가별 메달 수를 버블의 크기로 표현합니다.
  2.7. 국가 간 비교
  막대 그래프: 특정 연도나 전체 기간 동안의 국가별 메달 수를 비교합니다.
  히트맵: 국가와 메달 종류 간의 관계를 색상으로 표현합니다.
  트리맵: 국가별 메달 수를 면적으로 비교합니다.
  2.8. 추가적인 시각화 아이디어
  상관 행렬 히트맵: 변수들 간의 상관 관계를 시각화하여 중요한 인사이트를 도출합니다.
  트렌드 분석: 특정 국가나 종목의 메달 획득 트렌드를 시각화합니다.
  3. 활용 가능한 도구 및 라이브러리
  Python 라이브러리: Matplotlib, Seaborn, Plotly, Altair 등

  4. 시각화를 통해 얻을 수 있는 인사이트
  강국의 변화: 특정 국가가 특정 시기 이후 메달 획득에서 성장하거나 감소하는 패턴을 파악할 수 있습니다.
  종목별 경쟁도: 인기 있는 종목과 경쟁이 치열한 종목을 구분할 수 있습니다.
  선수 특성 분석: 선수의 연령, 성별 등이 메달 획득과 어떤 관련이 있는지 분석할 수 있습니다.

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````python
athletes_df = pd.read_csv('/content/athlete_events.csv')

regions_df = pd.read_csv('/content/noc_regions.csv')
````

````python
# lets merge athlete events and region based on NOC
df_merged = athletes_df.merge(regions_df,on='NOC',how='left')
````

````python
# Year 컬럼이 1968년 이상인 데이터만 필터링
df_filtered = df_merged[df_merged['Year'] >= 1968]
````

````python
# 각 종목(Sport)별로 평균 신장(Height)과 체중(Weight) 계산
avg_values = df_filtered.groupby('Sport')[['Height', 'Weight']].mean()

# 결측치를 채우는 함수 정의 (정수형으로 변환)
def fill_missing_values(row):
    if pd.isnull(row['Height']):
        row['Height'] = int(round(avg_values.loc[row['Sport'], 'Height']))  # 평균을 반올림 후 정수형으로 변환
    if pd.isnull(row['Weight']):
        row['Weight'] = int(round(avg_values.loc[row['Sport'], 'Weight']))  # 평균을 반올림 후 정수형으로 변환
    return row

# apply 함수로 결측치 채우기
df_filtered = df_filtered.apply(fill_missing_values, axis=1)
````
